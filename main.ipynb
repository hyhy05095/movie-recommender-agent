```python
import openai

client = openai.OpenAI()
messages = []
```

```python
PROMPT = """

You are a professional movie recommender.

Follow this conversation logic:

- If the user has not provided a favorite movie genre yet,
   ask: "Hello! Please tell me your favorite movie genre?"

-  If the user has provided their favorite genre but has not yet provided
   their favorite movie in that genre, then ask:
   "Which one is the best movie you have ever watched in that genre?"

- If the user has provided both their favorite genre and their favorite movie
   in that genre, then:
   - Recommend movies in that genre,
   - Avoid any movies the user has already watched,
   - Provide an expected rating (1â€“10) for each recommendation,
   - Keep your responses brief and focused.

"""

messages = [
    {"role": "system", "content": PROMPT}
]
```

```python
def call_ai():
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages = messages,
    )

    message = response.choices[0].message.content
    messages.append({"role": "assistant", "content": message})      
    print(f"AI: {message}")
```

```python
call_ai()
while True:
    message = input("You: ")    
    if message == "quit" or message == "q":
        break
    else:
        messages.append({"role": "user", "content": message})
        print(f"User: {message}")
        call_ai()
```
